Identificar las Provincias Más Productivas:

```python

# Agrupar por zona y calcular la suma de la cartera y las primas
productividad_zona = df.groupby('ZONA').agg({
    'CARTERA ACT.': 'sum',
    'PRIMAS NP.ANUALIZ. ACT.': 'sum'
}).reset_index()

# Renombrar columnas para claridad
productividad_zona.columns = ['Zona', 'Total_Cartera_Actual', 'Total_Primas_Actuales']

# Ordenar por total de cartera y primas en orden descendente
productividad_zona = productividad_zona.sort_values(by=['Total_Cartera_Actual', 'Total_Primas_Actuales'], ascending=False)

print(productividad_zona)
```

 Visualizar los Resultados:

```python

# Gráfico de barras para visualizar la cartera actual por zona
plt.figure(figsize=(12, 6))
plt.bar(productividad_zona['Zona'], productividad_zona['Total_Cartera_Actual'], color='blue', alpha=0.7)
plt.xlabel('Zona')
plt.ylabel('Total Cartera Actual')
plt.title('Cartera Actual por Zona')
plt.xticks(rotation=90)
plt.show()

# Gráfico de barras para visualizar las primas actuales por zona
plt.figure(figsize=(12, 6))
plt.bar(productividad_zona['Zona'], productividad_zona['Total_Primas_Actuales'], color='green', alpha=0.7)
plt.xlabel('Zona')
plt.ylabel('Total Primas Actuales')
plt.title('Primas Actuales por Zona')
plt.xticks(rotation=90)
plt.show()
```

Provincias con Mayor Siniestralidad:

```python
# Agrupar por zona y calcular el promedio de siniestralidad
siniestralidad_zona = df.groupby('ZONA').agg({
    '% SINIESTRALIDAD': 'mean'
}).reset_index()

# Renombrar columnas para claridad
siniestralidad_zona.columns = ['Zona', 'Promedio_Siniestralidad']

# Ordenar por siniestralidad promedio en orden descendente
siniestralidad_zona = siniestralidad_zona.sort_values(by='Promedio_Siniestralidad', ascending=False)

print(siniestralidad_zona)
```

```python
# Gráfico de barras para visualizar la siniestralidad promedio por zona
plt.figure(figsize=(12, 6))
plt.bar(siniestralidad_zona['Zona'], siniestralidad_zona['Promedio_Siniestralidad'], color='red', alpha=0.7)
plt.xlabel('Zona')
plt.ylabel('Promedio Siniestralidad (%)')
plt.title('Promedio de Siniestralidad por Zona')
plt.xticks(rotation=90)
plt.show()
```

-Total Primas Actuales: Las provincias con mayores sumas de primas actuales indican dónde se está recibiendo más dinero por la venta de nuevas pólizas.
